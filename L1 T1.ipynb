```python
df['Calories'].mean()
df['Calories'].median()
df['Calories'].std()
```

Average calories per item is X

Calories vary significantly across menu items

```python
df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
df.groupby('Category')['Calories'].mean().plot(kind='bar')
plt.title("Average Calories by Category")
plt.ylabel("Calories")
plt.show()
```

```python
top10 = df.sort_values('Calories', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10['Item'], top10['Calories'])
plt.xticks(rotation=45)
plt.title("Top 10 High Calorie Items")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Between Nutritional Values")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Correlation Between Nutritional Values")
plt.show()
```

### Correlation Analysis

The heatmap shows strong positive correlations between calories, total fat, and saturated fat.
This indicates that items higher in fat content tend to have higher calorie values.
Such insights can help in identifying high-calorie menu items and making healthier recommendations.
